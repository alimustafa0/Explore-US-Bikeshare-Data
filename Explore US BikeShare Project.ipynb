```python
import time
import pandas as pd
import numpy as np


CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }
months = ['all', 'january', 'february', 'march', 'april', 'may', 'june']
days = ['all', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    cities = ['chicago', 'new york city', 'washington']
    
    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    city = input("\nWould you like to see data for Chicago, New York City, or Washington?").lower()
    while city not in cities :
        print("Invalid input, Please choose amoung the list!")
        city = input("Again!!\nWould you like to see data for Chicago, New York City, or Washington?").lower()
        
    # TO DO: get user input for month (all, january, february, ... , june)
    month = input("\nEnter the name of the month that you want to analyze.\n(all, january, february, march, april, may, june)").lower()
    while month not in months :
        print("Invalid input, Please choose amoung the list!")
        month = input("Again!!\nEnter the name of the month that you want to analyze.\n(all, january, february, march, april, may, june)").lower()
        
    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    day = input("\nEnter the name of the day that you want to analyze.\n(all, monday, tuesday, wednesday, thursday, friday, saturday, sunday)").lower()
    while day not in days :
        print("Invalid input, Please choose amoung the list!")
        day = input("Again!!\nEnter the name of the day that you want to analyze.\n(all, monday, tuesday, wednesday, thursday, friday, saturday, sunday)").lower()

    print('-'*40)
    return city, month, day






def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    print("The data is loaded...")
    print("-" * 40)
    df = pd.read_csv(CITY_DATA[city])
    
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    
    df['month'] = df['Start Time'].dt.month
    
    df['day_of_week'] = df['Start Time'].dt.weekday_name
    
    df['hour'] = df['Start Time'].dt.hour
    
    if month != 'all':
        month = months.index(month)
        df = df[df['month'] == month]
        
    if day != 'all':
        df = df[df['day_of_week'] == day.title()]
        
    print('The data has been uploaded successfully.')
    
    return df






def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()
    
    input0 = input("Would you like to show you the most common name of the month, day and hour? (y/n)").lower()
    
    # TO DO: display the most common month
    while True:
        if (input0 == 'y'):
            popular_month = df['month'].mode()[0]
            print("\nThe most common month is " + str(months[popular_month]) + "[{}]".format(str(popular_month)))
    
    # TO DO: display the most common day of week
            popular_day = df['day_of_week'].mode()[0]
            print("\nThe most common day is " + popular_day + "[{}]".format(str(days.index(popular_day.lower()))))
    
    # TO DO: display the most common start hour
    
            popular_hour = df['hour'].mode()[0]
            print("\nThe most common hour is " + str(popular_hour))
            break
        elif input0 == 'n':
            print("I will skip this step.")
            break
        else:
            print('Invalid Value')
            input0 = input("Again!!\nWould you like to show you the most common name of the month, day and hour? (y/n)").lower()
    

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
    
    
    

    
    
    
def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()
    input0 = input("Would you like to show you the most common name of start and end station? (y/n)").lower()
    

    while True:
        if (input0 == 'y'):
            # TO DO: display most commonly used start station
            popular_start_st = df['Start Station'].mode()[0]
            print("\nMost commonly used start station is " + popular_start_st)
    
            
            # TO DO: display most commonly used end station
            popular_end_st = df['End Station'].mode()[0]
            print("\nMost commonly used end station is " + popular_end_st)
    

    
            Combination_Station = df.groupby(['Start Station', 'End Station'])[['Start Station', 'End Station']].count().max()
            print('\nMost Commonly used combination of start station and end station trip: ' +  popular_start_st + " & " + popular_end_st)
            print(str(Combination_Station))
            break
        elif input0 == 'n':
            print("I will skip this step.")
            break
        else:
            print('Invalid Value')
            input0 = input("Would you like to show you the most common name of start and end station? (y/n)").lower()

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

    
    
    
    
    
def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""
     #TO DO: display total travel time
    print('\nCalculating Trip Duration...\n')
    start_time = time.time()
    time0 = ['Minutes', 'Hours', 'Days']
    input0 = input("Do you want Trip Duration per Minutes, Hours or Days?").title()
    while True:
        if input0 == time0[0]: 
            Total_Travel_Time = sum(df['Trip Duration'])/60
            break
        elif input0 ==  time0[1]:
            Total_Travel_Time = sum(df['Trip Duration'])/3600
            break
        elif input0 ==  time0[2]:
            Total_Travel_Time = sum(df['Trip Duration'])/86400
            break
        else:
            print("Invalid input")
            input0 = input("Again!!\nDo you want Trip Duration per Minutes , Hours or Days?").title()
    print('\n\nTotal travel time: ' + str(Total_Travel_Time) + ' {}'.format(input0))
    # TO DO: display mean travel time
    Mean_Travel_Time = df['Trip Duration'].mean()
    print('\nMean travel time:' + str(Mean_Travel_Time/60) + " Minutes")

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
    
    
    
def user_stats(df):
    """Displays statistics on bikeshare users."""
    
    start_time = time.time()
    print('\nCalculating User Stats...\n')
    
    while True:
        input0 = input('Would you like to show you counts of user types and gender? (y/n)').lower()
        if input0 == 'y':
    # TO DO: Display counts of user types
            print('\n', df['User Type'].value_counts())
    # TO DO: Display counts of gender
            try:
                gender_type = df['Gender'].value_counts()
                print('\nGender Types are\n', gender_type)
            except:
                print("\nGender Types:\nNo data available for this month.")
            break
        elif input0 == 'n':
            print('I will skip this step.')
            break
        else:
            print('Invalid input.')
            input0 = input('Again!!\nWould you like to show you counts of user types and gender? (y/n)').lower()
    # TO DO: Display earliest, most recent, and most common year of birth
    while True:
        input1 = input('Would you like to show you earliest, most recent, and most common year of birth (y/n)')
        if input1 == 'y':
            try:
                earliest_year = df['Birth Year'].min()
                print('\nEarliest Year is', int(earliest_year))
            except:
                print("\nEarliest Year:\nNo data available for this month.")

            try:
                most_recent_year = df['Birth Year'].max()
                print('\nMost Recent Year is', int(most_recent_year))
            except:
                print("\nMost Recent Year:\nNo data available for this month.")

            try:
                most_common_year = df['Birth Year'].mode()
                print('\nMost Common Year is', int(most_common_year))
            except:
                print("\nMost Common Year:\nNo data available for this month.")
            break
        elif input1 == 'n':
                print('I will skip this step.')
                break
        else:
                print('Invalid input.')
                input1 = input('Again!!\nWould you like to show you earliest, most recent, and most common year of birth (y/n)')
                
        

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def main():

    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
	main()
```
